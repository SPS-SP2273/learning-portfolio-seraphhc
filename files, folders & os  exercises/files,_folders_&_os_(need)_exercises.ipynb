```python
def rename_my_file(old_file_name):
    text = old_file_name
    text_date = text.split('_')[0]  # this is oct-08. split oct-08 and singapore 
    text_city = text.split('_')[1] # thsi is singapore. split oct-08 and singapore 
    
    text_month = text_date.split('-')[0] # this is oct. split oct from everything
    text_day = text_date.split('-')[1] # this is 08
    
    if text_month in months:
        text_month = months[text_month]
    
    final = text_month + '-' + text_day + '-' + text_city # this is to ensure that there wil be no spaces in the final output
    return final
```

```python
def copy_files_to_folder(files, destination_folder):
    for file in files:
        shutil.copy(file, destination_folder)
        
cairo_files = glob.glob('os-collaboration-exercise-data/*cairo*')
destination_folder = 'Cairo'
copy_files_to_folder(cairo_files, destination_folder)
print ('files copied successfully')
```

```python
def move_files_to_folder(files, destination_folder):
    for file in files:
        shutil.move(file, destination_folder)
        
for city in ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila','Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']:
    city_file_path = "os-collabaration-exercise-data/*" + city.lower() + "*"
    city_files = glob.glob(city_file_path)
    print(city_file_path)
    destination_folder = os.path.join('cities', city)
    print(destination_folder)
    move_files_to_folder(city_files, destination_folder)
print('done')
```
