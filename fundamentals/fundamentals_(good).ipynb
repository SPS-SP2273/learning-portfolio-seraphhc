# There is more to if
to add more branches in an if statement, we can use elseif 

```python
name = 'Batman'

if name == 'Batman':
    print('Hello Batman!')
elif name == 'Robin':
    print('Hello Sidekick!')
else:
    print('Hello World!')
```

# Python stores information in different formats or types

```python
x = int(1.234)     #int() means that it will be stored as an integer (no decimal place)
print(x, type(x))
```

```python
x = str(1.234)   # str(): stored as an eng word
print(x, type(x))
```

```python
x = float(1.234)   #float() as a decimal number 
print(x, type(x))
```

```python
x = complex(1.234)  #complex() as a complex number (includes imaginary part of the no.)
print(x, type(x))
```

# Never compare floats directly
as there will be roundoff errors

```python
a = 0.1
a3 = 0.3
a * 3 == a3        #should be true!
```

```python
print(a * 3)
```

```python
f'{0.3:.17f}' #printing 0.3 to 17 decimal places
```

## A solution 
check if the variable is close to the expected values instead of checking for equality

```python
eps = 1E-10           #method 1 
abs(a * 3 - a3) < eps
```

```python
import numpy as np
```

```python
np.isclose(a * 3, a3)   #method 2
```

# Combining English (strings) and variables
This will be useful for example when automating personalized emails to a large class.

```python
name = "Batman"
print(f"Hello {name}!")
```

```python
name = "Batman"
print(f"Hello {name.upper()}!")
```

```python
x = 10
print(f"The value of {x} squared is {x**2}!")   #f-string / string interpolation 
```

```python
text = 'Bruce Wayne is Batman.'
print(f'{text}')        #f-string used to format text (string)
```

```python
print(f'{text:>30}')      # A block of 30 characters;
                          # aligned right
```

```python
print(f'{text:>25}')      # A block of 30 characters;
                          # aligned right
```

```python
print(f'{text:^30}')      # A block of 30 characters;
                          # aligned centre
```

```python
print(f'{text:<30}')      # A block of 30 characters;
                          # aligned left
```

```python
print(f'The cube of pi to 6 decimal places is {np.pi**3:.6f}') #f-string to format numbers
```

```python
print(f'The cube of pi to 6 decimal places is {np.pi**3:.6e}')
```
