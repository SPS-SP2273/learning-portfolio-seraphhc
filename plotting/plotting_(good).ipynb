# 3 What is OO ax, really?

```python
fig, ax = plt.subplots(nrows=2, ncols=1) # nrows gives us how many rows  
                                         # ncols gives us how many columns 
                                         # this gives us 2 axes 
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

#  A complete OO example


```python
fig, ax = plt.subplots(nrows=2, ncols=1,  # 2 rows 1 column 
                       figsize=(5, 5),   # changing figure size
                       sharex=True)      # ask the plots to share the x axis 

ax[0].plot(x, cos_x, label=r'$\cos(x)$')  #since ax is a numpy array w 2 axes
ax[0].fill_between(x, 0, cos_x, alpha=.25)  #we can index them using 0 and 1 
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)  #to fill the plots

for a in ax:        #instead of plotting 1 axis at a time we can use a for loop 
    a.legend()   
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')  #since we are sharing the x axis, only need to label the lowest plot 

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

fig.tight_layout()  #ask matplotlib to make adjustments to the layout to make the plot look nice
```

# 5 Multiple rows and columns

### Using ax

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten() #takes the 2D array and flattens it into a 1D array
# of can do the following: 
# top_left=ax[0, 0]
# top_right=ax[0, 1]
# bottom_left=ax[1, 0]
# bottom_right=ax[1, 1]

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')
# or can also do the following:
# ax[0, 0].set_title('Top Left')
# ax[0, 1].set_title('Top Right')
# ax[1, 0].set_title('Bottom Left')
# ax[1, 1].set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)    # to draw the lines
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

## 5.1 Sharing axes

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5)       #if we dont specify sharing of the x axis, matplotlib auto scaled the axes
)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row' #all the plots will have the same range for x axis
)
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

##  Histograms

### Data

### A quick helper function
create a function called det_plot_data() to allow me to read the data from these files serveral times 

```python
def get_plot_data():
    data = {} 
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1) #skip the first line as the first line in the data is the heading (words)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

### The histogram

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

##  Scatter plots
created by putting a markert at an (x,y) point we specify

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))  # to generate a list [0,1,2,3...]
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

##  Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers)   #using barh() will give horizontal bars instead
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

##  Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )
plt.title('Percentage of each major')
```
