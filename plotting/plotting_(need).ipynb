# A simple plot

## Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',       #this plots the lines of the graph, 
         color='red', linestyle='dashed')  #the order of how to specofy the keyword arguments do not matter 
   # label will reflect the name of the line in legend 
plt.xlabel('x-values') # these label the name of the x and y axis 
plt.ylabel('y-values')
plt.title('X vs Y') # this is the title of the graph 
plt.grid(alpha=.8) # this draws out the grids, with the number indicating the intensity of the colour of the grids
plt.legend(loc='upper right') #this specifies where the legend is placed on the graph
```

some can be abbreviated:
- colour -> c
- linestyle -> ls
- linewidth -> lw 

```python
plt.plot(x, y_1, y_2) # this results in a plot that had 2 lines 
```

## Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.plot(x, y_2,
         label='Y2 values', color='blue')  # this line is added
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

##  Yet another plot but with error bars

to include x and y error bars for the points, we can use the plotting command errorbar()

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]   # new additions 
x_error, y_error = 0.1, 0.75       # the 'width', 'height' of error bars

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')   #new addition 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# Better with NumPy

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
```

##  Adding mathematical functions
numpy allows us to generate data-related math functions easily 

```python
# to plot x^2 and sin(x)

x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()      
plt.grid(alpha=.25)
```

the above plot does not look good since sin(x) is between +/-1 , but x^2 has no bounds. thus... we can add ano y axis sharing the same x-axis 

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()            # This creates a new y-axis for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

the above plot does not look good as we only have a few plots

```python
x = np.linspace(0, 6, 100) # here !
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  

plt.twinx()           
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 
```

# Saving to disc
to use a plot in a report or presentation, we must save it in a disk using **savefig()**

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150) # ('filename of choice.extension', resolution using dots per inch)
```

we can also specify the path:  plt.savefig('~/Desktop/simple-plot.png', dpi=150) 

to obtain jpeg format, we can use **simple-plot.jpeg**

# A real example: Global Warming

##  Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0] # the dates were the first elements in the file
anomaly = data[:, -1] #the anomaly were the last elements in the file 

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])  #this limits the y axis
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

# adding a horizontal line at the 0 vale to highlight trend of data 
# hlines() need the y value and staring and ending values for x 
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## Styles
make plots look decent without spending too much time customising it

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.xlabel('date')
plt.ylabel('temp. anomaly')
plt.title('temp anomaly over the years')

plt.style.use('fivethirtyeight')
```
